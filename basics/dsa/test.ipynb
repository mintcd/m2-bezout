```python
def _gray_word(i, L):
    g = i ^ (i >> 1)
    return ''.join('B' if (g >> k) & 1 else 'A' for k in reversed(range(L)))

def worst_case_u_gray(n):
    if n <= 0: return ""
    # largest power of two strictly less than n
    L = 1
    while 2 * L < n:
        L *= 2
    if L >= n:  # safety
        L //= 2
    L = max(L, 1)

    full = n // L
    rem  = n - full * L

    start = 1  # skip the all-zero Gray word
    words = [_gray_word(start + t, L) for t in range(full)]
    if rem:
        words.append(_gray_word(start + full, L)[-rem:])  # use tail, not head
    return ''.join(words)
```
